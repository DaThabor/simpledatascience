## Correlation

<br>

Correlation means that 2 variables x and y are linked to each other when the correlation is either high (+1) or low (-1) and not correlated at all (+-0) if there is no link between the two.

See below examples of correlation as:

- Full Positive correlation, when x increases, y increases at the same speed
- Full negative correlation, when x increases, y decreases at the same speed
- No correlation, when there is no correlation between the two.

<br>

----------

### Positive correlation

When the value of x increases, the value of y increases at the same speed. Three variants can be found here:


1. Perfect positive correlation (correlation == 1)
2. High positive correlation (correlation 0.5 < x < 1
3. Low positive correlation (correlation ~ 0.5)

<br>

#### Perfect positive correlation


```python
x_perfect = [0,1,2,3,4,5]
y_perfect = [0,1,2,3,4,5]

df_perfect = pd.DataFrame({'x':x_perfect, 'y':y_perfect})

plot_perfect_1 = df_perfect.plot('x', 'y', kind='scatter')
perfect_1_corr = df_perfect.corr()

plt.show()
```

```python
plot_perfect_corr = sns.heatmap(df_perfect.corr(), 
                        cmap="YlGnBu", 
                        annot = True, 
                        linewidths=.5)

plt.show()
```

#### High Positive correlation

```python
x_high = [0,1,2,3,4,5]
y_high = [0.2,2.4,1.6,3.6,3.2,5.4]

df_high = pd.DataFrame({'x':x_high, 'y':y_high})

plot_perfect_2 = df_high.plot('x', 'y', kind='scatter')
perfect_2_corr = df_high.corr()

plt.show()
```

```python
plot_high_corr = sns.heatmap(df_high.corr(), 
                        cmap="YlGnBu", 
                        annot = True, 
                        linewidths=.5)

plt.show()
```

#### Low positive correlation

```python
x_low = [0,1,2,3,4,5]
y_low = [0.4,3.1,3.8,1.8,1.7,4.8]

df_low = pd.DataFrame({'x':x_low, 'y':y_low})

plot_perfect_3 = df_low.plot('x', 'y', kind='scatter')
perfect_3_corr = df_low.corr()

plt.show()
```

```python
plot_low_corr = sns.heatmap(df_low.corr(), 
                        cmap="YlGnBu", 
                        annot = True, 
                        linewidths=.5)

plt.show()
```
